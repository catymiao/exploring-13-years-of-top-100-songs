```python
#df[]
import numpy as np
df['lyrics'] = np.where(df['lyrics'].isna(), '', df['lyrics'])
```

```python
# https://stackoverflow.com/a/47091490/4084039
import re

def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

```python
from nltk.corpus import stopwords
stopwords=list(stopwords.words('english'))
stopwords[0:10]
```

```python
# import string
# input_str = 'This &is [an] example? {of} string. with.? punctuation!!!!” # Sample string'
# translator = str.maketrans('', '', string.punctuation)
# text=input_str.translate(translator) 
# text

from nltk.stem import WordNetLemmatizer 
from nltk.tokenize import word_tokenize 
lemmatizer = WordNetLemmatizer() 
# lemmatize string 
def lemmatize_word(text, output = 'as_string'): 
    word_tokens = word_tokenize(text) 
    # provide context i.e. part-of-speech 
    if output == 'as_list':
        lemmas = [lemmatizer.lemmatize(word, pos ='v') for word in word_tokens] 
    elif output == 'as_string':
        lemmas = [lemmatizer.lemmatize(word, pos ='v') for word in word_tokens]
        lemmas = ' '.join(lemmas)
    return lemmas 
```

```python
# Combining all the above stundents 
from tqdm import tqdm
preprocessed_lyrics = []
# tqdm is for printing the status bar
for lyric in tqdm(df['lyrics'].values):
    #if len(str(lyric)) != 0:
        #print (lyric)
    sent = decontracted(lyric)

    #remove numbers
    sent = re.sub(r'\d+', '', sent)
    #remove puncutuation and special characters
    translator = str.maketrans('', '', string.punctuation)
    sent=sent.translate(translator) 

    sent = sent.replace('\\r', ' ')
    sent = sent.replace('\\"', ' ')

    #remove special words in the songs
    sent = sent.replace('[Interlude]', '')
    sent = sent.replace('[Refrain]', '')
    sent = sent.replace('[Outro]', '')
    sent = sent.replace('[Chorus]', '')
    sent = sent.replace('Verse', '')
    sent = sent.replace('\\n', ' ')
    sent = sent.replace('b\d+\b', '')
    #sent = sent.replace('[', '')
    #sent = sent.replace(']', '')
    # https://gist.github.com/sebleier/554280
    sent = ' '.join(e for e in sent.split() if e.lower() not in stopwords)
    preprocessed_lyrics.append(sent.lower().strip())
    #sent = lyric

df['lyrics_clean'] = preprocessed_lyrics
```
